## SQL Database


```python
#create sample sqlite database
import sqlite3
import os 

os.makedirs("data/databases", exist_ok= True)
```

```python
conn = sqlite3.connect("data/databases/company.db")
cursor = conn.cursor()
```

```python
cursor.execute('''CREATE TABLE if not exists company (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, description TEXT)''')
```

```python
cursor.execute('''CREATE TABLE if not exists employees (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, description TEXT)''')
```
