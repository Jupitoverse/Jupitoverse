### Load PDF Files

```python
from langchain_community.document_loaders import (
    PyPDFLoader,
    PyMuPDFLoader,
    UnstructuredPDFLoader
)
```

```python
print("PyPDFLoader")

try:
    pypdf_loader = PyPDFLoader("data/pdf/Rebill.pdf")
    docs = pypdf_loader.load()
    print(f"Number of documents: {len(docs)}")
    print("------------------------")
    print("First document:")
    print("------------------------")
    print(docs[0].page_content[:100])
    print("------------------------")
    print(docs[0].metadata)
    print("------------------------")
    print("PDF Content")
    print("------------------------")
    print(docs)
except Exception as e:
    print(f"Error loading PyPDFLoader: {e}")
    
```

 Data Format

Type: Python objects (likely from LangChain or similar RAG framework).
Underlying format:

List → contains multiple Document objects.
Each Document → has:

metadata (dictionary)
page_content (string)






✅ Data Structure Explanation for Interview
You can explain it like this:

“The data returned after loading a PDF in a RAG pipeline is typically a list of Document objects. Each document represents one page or chunk of text from the PDF. The structure is hierarchical:

At the top level, we have a list because a PDF can have multiple pages.
Each element in the list is a Document object, which contains:

metadata → a dictionary storing attributes like author, title, creation date, source file path, page number, etc.
page_content → a string containing the actual text extracted from that page.



This structure is ideal for RAG because metadata helps in filtering and retrieval, while page_content is used for embedding and semantic search.”


✅ Key Points to Mention in Interview

Why this structure?

Easy to chunk and index for vector databases.
Metadata supports filtering and contextual retrieval.


Common use in RAG:

Convert page_content into embeddings.
Store metadata for query-based filtering.


Serialization:

Can be converted to JSON for storage or API transfer.
